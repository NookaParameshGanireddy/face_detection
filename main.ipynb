```python
import cv2

# Load image
image = cv2.imread("C:/Users/AiUnika/Desktop/blur_faces/dh-number-plate-car-hong-kong-bmw-x5-personal-number-plate-hong-kong-EPWTMY.jpg")

# Load pre-trained classifiers
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
plate_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml')

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces and plates
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
plates = plate_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=4)

# Function to blur region
def blur_region(img, x, y, w, h):
    roi = img[y:y+h, x:x+w]
    roi = cv2.GaussianBlur(roi, (51, 51), 30)
    img[y:y+h, x:x+w] = roi
    return img

# Blur detected regions
for (x, y, w, h) in faces:
    image = blur_region(image, x, y, w, h)

for (x, y, w, h) in plates:
    image = blur_region(image, x, y, w, h)

# Save or display result
cv2.imwrite('blurred_output.jpg', image)
# cv2.imshow('Blurred Image', image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
import cv2

# Load image
image = cv2.imread('C:/Users/AiUnika/Desktop/blur_faces/Latest-car-number-plate-designs.jpg')
if image is None:
    print("Image not found. Check file path.")
    exit()

# Load Haar cascades
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
plate_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml')

if face_cascade.empty() or plate_cascade.empty():
    print("Failed to load Haar cascade classifiers.")
    exit()

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces and plates
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3)
plates = plate_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3)

print(f"Detected {len(faces)} faces")
print(f"Detected {len(plates)} number plates")

# Blur function
def blur_region(img, x, y, w, h):
    roi = img[y:y+h, x:x+w]
    if roi.size == 0:
        return img  # skip invalid region
    roi = cv2.GaussianBlur(roi, (51, 51), 30)
    img[y:y+h, x:x+w] = roi
    return img

# Apply blur
for (x, y, w, h) in faces:
    image = blur_region(image, x, y, w, h)

for (x, y, w, h) in plates:
    image = blur_region(image, x, y, w, h)

# Save output
cv2.imwrite('blurred_output.jpg', image)
print("Saved blurred_output.jpg")
```

```python
import cv2
from ultralytics import YOLO

# Load YOLO models
face_model = YOLO("yolov8n-face.pt")  # Custom trained face model (you can train or download one)
plate_model = YOLO("license_plate_detector.pt")  # Replace with path to number plate model

# Load your input image
image_path = "C:/Users/AiUnika/Desktop/blur_faces/OIP.jpg"
img = cv2.imread(image_path)

# Detect faces
results_face = face_model(img)
for r in results_face:
    for box in r.boxes.xyxy:
        x1, y1, x2, y2 = map(int, box)
        roi = img[y1:y2, x1:x2]
        roi = cv2.GaussianBlur(roi, (51, 51), 30)
        img[y1:y2, x1:x2] = roi

# Detect number plates
results_plate = plate_model(img)
for r in results_plate:
    for box in r.boxes.xyxy:
        x1, y1, x2, y2 = map(int, box)
        roi = img[y1:y2, x1:x2]
        roi = cv2.GaussianBlur(roi, (51, 51), 30)
        img[y1:y2, x1:x2] = roi

# Save or display the result
cv2.imwrite("blurred_output.jpg", img)
# cv2.imshow("Blurred", img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
